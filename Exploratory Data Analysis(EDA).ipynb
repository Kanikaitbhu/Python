## Variables transformations

```python
from datetime import datetime
Now = datetime.now()
```

## Year_Birth
converting the age in birth year to age in number of years of age.
Also replacing the outliers (Age>100) with the mean age.

```python
#converting the year to Age in years
cust_data["Age"]=(2021- cust_data["Year_Birth"])
```

```python
cust_data["Age"] 
```

```python
#plotting the distribution of age to see if any outliers are present.
sns.boxplot(cust_data["Age"])
```

```python
## replacing the outliers (>100 values) with the mean age
cust_data["Age"] = np.where(cust_data["Age"]> 100, cust_data["Age"].mean(), cust_data["Age"])
```

```python
# plotting again to check the updated data
sns.boxplot(cust_data["Age"])
```

## Dtcustomer
converting the joining date to the year joined.

```python
## converting the dates in inconsitent formats to standard single format.
cust_data['Dt_Customer'] = pd.to_datetime(cust_data['Dt_Customer'])
```

```python
cust_data["Age_Joined"]= cust_data['Dt_Customer'].dt.year-cust_data["Year_Birth"]
```

```python
cust_data['Dt_Customer'].dt.year
```

```python
## checking the distribution of the joining days
sns.countplot(cust_data["Dt_Customer"].dt.year)
```

## Income
Replacing the income outlier value>600000 with the mean value

```python
cust_data["Income"] = np.where(cust_data["Income"]> 600000, cust_data["Income"].mean(), cust_data["Income"])
```

for the categorical variables Education, Marital_status and country we have to do encoding.

```python
cust_data["Marital_Status"].unique()
```

```python
#the three categories YOLO, Alone and Absurd can be combined together to others.
cust_data["Marital_Status"] = np.where(cust_data["Marital_Status"]== "YOLO", "Others", cust_data["Marital_Status"])
cust_data["Marital_Status"] = np.where(cust_data["Marital_Status"]== "Alone", "Others", cust_data["Marital_Status"])
cust_data["Marital_Status"] = np.where(cust_data["Marital_Status"]== "Absurd", "Others", cust_data["Marital_Status"])
```

```python
# Creating levels for the marital status
marital_status_dic = {"Single": 0,"Married": 1,"Divorced": 2, "Together" :3, "Widow" :4, "Others" :5 }
cust_data["Marital_Status"] = cust_data.Marital_Status.map(marital_status_dic) 
```

```python
cust_data["Education"].unique()
```

```python
# creating levels for Education
Education_status_dic = {"Graduation": 0,"PhD": 1,"2n Cycle": 2, "Master" :3, "Basic" :4 }
cust_data["Education"] = cust_data.Education.map(Education_status_dic) 
```

### Country level analysis

```python
# Country Mean and Median inccome using pivot function

df = pd.pivot_table(cust_data, 
                     values='Income', 
                     index='Country', 
                     aggfunc={'Income':['mean','median']})
df.plot(kind='bar')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.ylabel('Income')
plt.show()
```

ME has the highest Mean income. US has higher Median income than Mean income.

```python
cust_data['Country'].value_counts()
```

```python
# Plotting the Campaign effectiveness in each country

df = pd.pivot_table(cust_data, 
                     values= ['Response','AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5'],
                     index='Country', 
                     aggfunc={'Response':'mean','AcceptedCmp1':'mean','AcceptedCmp2':'mean','AcceptedCmp3':'mean',
                              'AcceptedCmp4':'mean','AcceptedCmp5':'mean' })

df.plot(kind = 'bar')
plt.legend(bbox_to_anchor = (1.05,1),loc = 'upper left')
plt.ylabel('Offer Accepted')
plt.xlabel('Country')
plt.show()
```

```python
## Heat map for the campain effectiveness
Campaigns_category = cust_data.loc[:,['Country','Response','AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']]
temp = Campaigns_category.groupby(by='Country').agg(['mean'])

temp.drop(['ME'], axis=0, inplace=True)  # since "ME" have only 3 observations, I decided to drop it in this table.
temp.style.background_gradient(sns.light_palette('green', as_cmap=True))
```

From the above Heat map we can say that Campaign 1 was effective in SP,Campaign 2 was not very effective, campaign 3 was effective in  GER and IND, Campaign 4 was effective in GER and Campaign 5 was effective in SP.

```python
## Plotting the mean Purchase in each category for each country
df= pd.pivot_table(cust_data, 
             values = ['NumDealsPurchases',"NumWebPurchases",'NumCatalogPurchases', 'NumStorePurchases','NumWebVisitsMonth'],
                   index = 'Country',
                  aggfunc={'NumDealsPurchases':'mean',"NumWebPurchases":'mean','NumCatalogPurchases':'mean', 
                          'NumStorePurchases':'mean','NumWebVisitsMonth':'mean'})
df.plot(kind='bar')
plt.legend(bbox_to_anchor = (1.05,1),loc = 'upper left')
plt.ylabel('Average purchases')
plt.show()
```

```python
df.drop(["ME"],axis=0,inplace=True)  # since "ME" have only 3 observations, I decided to drop it in this table.
df.style.background_gradient(sns.light_palette('green', as_cmap=True))
```

US had highest average numbers in all the Purchase category.
